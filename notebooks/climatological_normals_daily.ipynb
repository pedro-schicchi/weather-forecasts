#### Imports

```python
roi = ee.Geometry.Polygon([[
    [-179, 73],
    [-179, -60],
    [180, -60],
    [180, 73]
]], None, False);
```

```python
# era5 = ee.ImageCollection("ECMWF/ERA5_LAND/DAILY_AGGR") \
era5 = ee.ImageCollection("ECMWF/ERA5/DAILY") \
    .filter(ee.Filter.calendarRange(years[0], years[1], 'year')) \
    .select(['mean_2m_air_temperature', 'total_precipitation'])
```

```python
# loop over days of year and extract means
for doy in range(1,366):
    # filter that day of year
    era5_filtered = era5.filter(ee.Filter.calendarRange(doy, doy, 'day_of_year'))
    
    # export mean
    geemap.ee_export_image(
        era5_filtered.mean(),
        filename=os.path.abspath(f'../data/raw/era5_{years[0]}-{years[1]}_doy_{doy}_mean.tif'),
        region=roi,
        file_per_band=False
    )
    
    # export std
    geemap.ee_export_image(
        era5_filtered.reduce(ee.Reducer.stdDev()),
        filename=os.path.abspath(f'../data/raw/era5_{years[0]}-{years[1]}_doy_{doy}_std.tif'),
        region=roi,
        file_per_band=False
    )
```
